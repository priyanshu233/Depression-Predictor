```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/depression_data.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df_s1 = pd.DataFrame(df['Dietary Habits'].value_counts())
```

```python
df_s1.head()
```

```python
df_s1.rename_axis('Dietary Habits Type',inplace=True)
df_s1.columns = ['Total Count']
```

```python
df_s1 = df_s1.reset_index()
```

```python
df_s1
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data = df_s1,x='Dietary Habits Type',y='Total Count',edgecolor='yellow')
plt.title('Dietary Habits overview on depression',fontsize=18,weight='bold')
plt.xlabel('Dietary Habits Type',fontsize=14,weight='bold')
plt.ylabel('Depressed People',fontsize=14,weight='bold')

for i in range(len(df_s1)):
  plt.text(i,df_s1['Total Count'][i],df_s1['Total Count'][i],ha='center',va='bottom',fontsize=12,fontweight='bold')


plt.show()
```

```python
smoking_status = pd.DataFrame(df['Smoking Status'].value_counts())
smoking_status.columns = ['Total_count']
smoking_status = smoking_status.reset_index()
smoking_status.head()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data = smoking_status,x='Smoking Status',y='Total_count',edgecolor='yellow')
plt.title('Smoking Habits overview on depression',fontsize=18,weight='bold')
plt.xlabel('Smoking Status',fontsize=14,weight='bold')
plt.ylabel('Depressed People',fontsize=14,weight='bold')

for i in range(len(smoking_status)):
  plt.text(i,smoking_status['Total_count'][i],smoking_status['Total_count'][i],ha='center',va='bottom',fontsize=12,fontweight='bold')


plt.show()
```

```python
employement_status = pd.DataFrame(df['Employment Status'].value_counts())
employement_status.columns = ['Total_count']
employement_status = employement_status.reset_index()
employement_status.head()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data = employement_status,x='Employment Status',y='Total_count',edgecolor='yellow')
plt.title('Employment Status effects by depression',fontsize=18,weight='bold')
plt.xlabel('Employment Status',fontsize=14,weight='bold')
plt.ylabel('Depressed People',fontsize=14,weight='bold')

for i in range(len(employement_status)):
  plt.text(i,employement_status['Total_count'][i],employement_status['Total_count'][i],ha='center',va='bottom',fontsize=12,fontweight='bold')


plt.show()
```

```python
alcohol_consumption = pd.DataFrame(df['Alcohol Consumption'].value_counts())
alcohol_consumption.columns = ['Total_count']
alcohol_consumption = alcohol_consumption.reset_index()
alcohol_consumption.head()
```

```python
alcohol_consumption['Total_count'].plot(kind='pie',autopct='%0.1ff%%',colors=['#66b3ff','#99ff99','#ffcc99'],shadow=True,startangle=90,wedgeprops={'edgecolor':'black'})

plt.title('Alcohol consumption overview depression ',fontsize=15,weight='bold')
plt.axis('equal')
plt.show()
```
